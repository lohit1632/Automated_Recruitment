#Extracting the Degree of Candidate from CV and adding it to the dataframe

```python
import re
import os
import pandas as pd
from docx import Document
import zipfile
import os
import pandas as pd
from docx import Document
import shutil
import tempfile
def find_degrees(text):
    # Define regex patterns for degrees
    degree_patterns = [
        r"(B\.Tech|Bachelor of Technology)",
        r"(BA|Bachelor of Arts)",
        r"(BCom|Bachelor of Commerce)"
        # Add more patterns for other degrees as needed
    ]

    degrees_found = set()
    for pattern in degree_patterns:
        matches = re.findall(pattern, text, re.IGNORECASE)
        degrees_found.update(matches)

    return degrees_found
def search_languages(text):
    # Define language patterns
    language_patterns = [
    r"english",
    r"hindi",
    r"bengali",
    r"telugu",
    r"marathi",
    r"tamil",
    r"urdu",
    r"kannada",
    r"gujarati",
    r"odia",
    r"malayalam",
    r"punjabi",
    r"assamese",
    r"maithili",
    r"sanskrit",
    r"kashmiri",
    r"sindhi",
    # Add more patterns for other languages as needed
]

    languages_found = set()
    for pattern in language_patterns:
        matches = re.findall(pattern, text, re.IGNORECASE)
        languages_found.update(matches)

    return languages_found
def search_skills(text):
    # Define skill patterns
    skill_patterns = [
        r"sales",
        r"accounting",
        r"communication",
        r"computer",
        r"mathematics",
        # Add more patterns for other skills as needed
    ]

    skills_found = set()
    for pattern in skill_patterns:
        matches = re.findall(pattern, text, re.IGNORECASE)
        skills_found.update(matches)

    return skills_found

def extract_text_from_docx(docx_file):
    doc = Document(docx_file)
    full_text = []

    # Extract text from paragraphs
    for para in doc.paragraphs:
        full_text.append(para.text)

    # Extract text from tables
    for table in doc.tables:
        for row in table.rows:
            for cell in row.cells:
                full_text.append(cell.text)

    return '\n'.join(full_text)

train_data = pd.read_csv("train.csv")
folder_path = "/content/CV_train"

files = os.listdir(folder_path)

for index, row in train_data.iterrows():
    unique_id = row["CandidateID"]  # Assuming "CandidateID" is the name of the column containing the IDs

    # Construct the path to the resume file
    path = "/content/CV_train/"
    resume_path = os.path.join(path, f"{unique_id} Resume.docx")
    #print(unique_id)
    path2 = unique_id
    path2+=" Resume.docx"
    # Check if the resume file exists
    try :
        # Extract text from the resume file
        resume_text = extract_text_from_docx(resume_path)
        cv_text = resume_text.lower()

        # Find degrees in the CV text
        degrees = find_degrees(cv_text)

        train_data.loc[index, "Degree"] = ", ".join(degrees) if degrees else None


    except:
        print(f"Resume file not found for ID: {unique_id}")
        train_data.loc[index, "Degree"] = -1

train_data
```

#Extracting the Birth Year

```python

def search_birth_year(text):
    # Define regex pattern to find a 4-digit number within the range of 1950 to 2010

    pattern = r"\b(19[5-9]\d)\b"

    # Search for the pattern in the text
    match = re.search(pattern, text)

    # If a match is found, return the birth year
    if match:
        return match.group()

    return None

for index, row in train_data.iterrows():
    unique_id = row["CandidateID"]  # Assuming "CandidateID" is the name of the column containing the IDs

    # Construct the path to the resume file
    path = "/content/CV_train/"
    resume_path = os.path.join(path, f"{unique_id} Resume.docx")
    #print(unique_id)
    path2 = unique_id
    path2+=" Resume.docx"
    # Check if the resume file exists
    # Extract text from the resume file
    try:
      resume_text = extract_text_from_docx(resume_path)
      cv_text = resume_text.lower()
      date = search_birth_year(cv_text)
      train_data.loc[index, "Date of Birth"] = date if date else None

    except:
        print(f"Resume file not found for ID: {unique_id}")

train_data.head(10)
```

#Checking the areas in which the candidate has Experience

```python
from collections import Counter

# Assuming loan_types_column is a pandas Series containing text data
loan_types_column = data["Which Products you are selling in your pervious role ?"]

# Convert all non-string values to strings
loan_types_column = loan_types_column.astype(str)

# Combine the text into a single string
combined_text = ' '.join(loan_types_column)

# Tokenize the text
words = combined_text.split(';')

# Convert to lowercase to ensure consistency
words = [word.strip().lower() for word in words]

# Count occurrences of each word
word_counts = Counter(words)

# Get the top 7 most common words
top_words = word_counts.most_common(7)

# Filter out the word "other"
top_words_other_than_other = [word for word, count in top_words if word != 'others']

# Select the first word from the filtered list
top_word_other_than_other = top_words_other_than_other[0]

# Create new columns for each top word
for word in top_words_other_than_other:
    # Create a new column with the word as the column name
    if word != "others":
      data[word] = data["Which Products you are selling in your pervious role ?"].apply(
        lambda x: 1 if word in str(x).lower() else 0
    )

# Print the top occurring word other than "other"
print("Top occurring word other than 'other':", top_word_other_than_other)
data
```
